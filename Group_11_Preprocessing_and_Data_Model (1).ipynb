**Scatter Plot (Relationship Analysis)**

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, ax = plt.subplots(figsize=(14, 7))

telco["TotalCharges"] = pd.to_numeric(telco["TotalCharges"], errors='coerce')

churn_0 = telco[telco["Churn"] == 0]
churn_1 = telco[telco["Churn"] == 1]

# Plot No Churn with glow effect (double layer)
ax.scatter(churn_0["tenure"], churn_0["TotalCharges"],
           color='#00BFFF', alpha=0.15, s=80, edgecolors='none')  # glow
ax.scatter(churn_0["tenure"], churn_0["TotalCharges"],
           color='#007DC6', alpha=0.6, s=25,
           edgecolors='#004F8B', linewidth=0.3,
           label=f'Stayed  ({len(churn_0):,} customers)')

# Plot Churned with glow effect
ax.scatter(churn_1["tenure"], churn_1["TotalCharges"],
           color='#FF6B6B', alpha=0.2, s=100, edgecolors='none')  # glow
ax.scatter(churn_1["tenure"], churn_1["TotalCharges"],
           color='#E8000B', alpha=0.8, s=35,
           edgecolors='#8B0000', linewidth=0.4,
           label=f'Churned ({len(churn_1):,} customers)')

# Trend line for stayed customers
z = np.polyfit(churn_0["tenure"].dropna(),
               churn_0["TotalCharges"].dropna(), 1)
p = np.poly1d(z)
x_line = np.linspace(0, 72, 100)
ax.plot(x_line, p(x_line), '--', color='#004F8B',
        linewidth=1.8, alpha=0.7, label='Trend — Stayed')

# Divider zone
ax.axvspan(0, 15, alpha=0.06, color='red', label='High Risk Zone (0–15 months)')

# Titles and labels
ax.set_title('Tenure vs Total Charges — Churn Risk Analysis\n',
             fontsize=16, fontweight='bold', color='#1A1A2E')
ax.set_xlabel('Tenure (Months)', fontsize=13, labelpad=10)
ax.set_ylabel('Total Charges ($)', fontsize=13, labelpad=10)

# Annotations
ax.annotate('⚠  HIGH CHURN ZONE\nMost cancellations happen\nin first 15 months',
            xy=(7, 6500), fontsize=10, color='#8B0000', fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='#FFE8E8',
                      edgecolor='#E8000B', linewidth=1.5))

ax.annotate('✔  LOYAL CUSTOMERS\nLong tenure = high lifetime\nvalue — rarely churn',
            xy=(45, 7500), fontsize=10, color='#004F8B', fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='#E8F4FD',
                      edgecolor='#007DC6', linewidth=1.5))

# Stats box
churn_rate_early = len(telco[(telco["tenure"] <= 15) & (telco["Churn"] == 1)]) / \
                   len(telco[telco["tenure"] <= 15]) * 100
churn_rate_late  = len(telco[(telco["tenure"] >  45) & (telco["Churn"] == 1)]) / \
                   len(telco[telco["tenure"] >  45]) * 100

stats_text = f'Churn Rate (0–15 months):  {churn_rate_early:.1f}%\n' \
             f'Churn Rate (45+ months):    {churn_rate_late:.1f}%'
ax.text(0.98, 0.05, stats_text, transform=ax.transAxes,
        fontsize=10, verticalalignment='bottom', horizontalalignment='right',
        bbox=dict(boxstyle='round,pad=0.6', facecolor='#FFFDE7',
                  edgecolor='#FFC107', linewidth=1.5))

ax.legend(fontsize=11, loc='upper left', frameon=True,
          facecolor='white', edgecolor='#CCCCCC', framealpha=0.95)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_facecolor('#F8F9FA')
fig.patch.set_facecolor('#FFFFFF')

plt.tight_layout()
plt.show()
```

**Telco Customer Churn Dataset**

The correlation analysis for the Telco dataset shows meaningful relationships between customer characteristics and churn behavior.

Tenure shows a strong negative correlation with Churn, indicating that customers who stay longer are less likely to leave.

MonthlyCharges has a positive correlation with churn, suggesting that customers paying higher monthly fees are more likely to churn.

TotalCharges is strongly correlated with tenure, which is expected since total charges accumulate over time.

Service-related features (like contract type and internet service) show moderate correlations with churn.

📌 Interpretation:

Customer loyalty (tenure) is the strongest protective factor against churn. High monthly charges increase churn risk. This suggests retention strategies should focus on long-term engagement and pricing optimization.

**Model Parameters and Algorithm-Level Monitoring Under Drift**

Building upon the baseline established during the EDA phase, the next stage of the project will focus on evaluating how data and concept drift influence internal model behavior and algorithmic stability. While distributional changes in input features are important, true drift impact must be assessed at the model parameter level.

For parametric models such as Logistic Regression, drift may manifest through instability in learned coefficients (β values), particularly for high-impact features such as tenure, repayment history, and financial burden indicators. Significant changes in coefficient magnitude or direction over time may signal concept drift affecting decision boundaries.

For tree-based algorithms (e.g., Random Forest and Gradient Boosting), drift may alter:

* Feature importance rankings

* Split thresholds

* Tree structure and depth patterns

Changes in these parameters can indicate shifts in the model’s learned structure and predictive logic.

In neural network models, distributional shifts may impact:

* Weight magnitudes

* Bias terms

* Activation sensitivity

Output probability calibration

Even if overall accuracy remains stable, shifts in predicted probability distributions may signal declining model reliability.

Across all algorithms, the following parameters will be monitored to assess drift-related risk:

* Model coefficients and weight stability

* Feature importance scores

* Decision boundary behavior

* Predicted probability distributions

* Classification threshold sensitivity

Performance metrics (accuracy, precision, recall, F1-score, AUC)

By linking EDA findings to algorithm-level parameter monitoring, the project moves beyond descriptive analysis and establishes a framework for auditing model robustness. This ensures that deployed systems remain reliable, interpretable, and stable under evolving real-world data conditions.
